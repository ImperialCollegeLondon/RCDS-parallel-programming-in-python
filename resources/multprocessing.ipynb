```python
import matplotlib.pyplot as plt

n = [int(1e3), int(1e4), int(1e5), int(1e6), int(1e7), int(1e8), int(1e9), int(1e10)]

time_no_concurrency = [0.026, 0.037, 0.041, 0.19, 1.0, 9.5, 95.2, 1628]
time_1 = [0.25, 0.26, 0.27, 0.37, 1.68, 9.9, 99.7, 1902]
time_2 = [0.54, 0.38, 0.46, 0.84, 0.94, 6.9, 61.7, 956]
time_4 = [0.52, 0.67, 0.56, 0.61, 1.1, 4.5, 42.3, 465]
time_8 = [0.82, 0.83, 0.79, 0.86, 1.1, 3.6, 40, 408]

fig, ax = plt.subplots()

ax.plot(n, time_no_concurrency, label='No concurrency', linestyle='solid')
ax.plot(n, time_1, label='1 thread', linestyle='--')
ax.plot(n, time_2, label='2 threads', linestyle='-.')
ax.plot(n, time_4, label='4 threads', linestyle = '-')
ax.plot(n, time_8, label='8 threads', linestyle=':')

ax.set(xlabel='n', ylabel='Time (s)')

ax.legend()

ax.set_yscale('log')
ax.set_xscale('log')

fig.savefig("queue_smallest_factor.png")
```
