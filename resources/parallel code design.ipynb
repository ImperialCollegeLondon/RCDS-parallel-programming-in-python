```python
import numpy as np
import matplotlib.pyplot as plt

def amdahl(P, n):
    return 1 / ((1 - P) + P / n)

n_cores = 2 ** np.arange(1, 15)

print(n_cores)

speed_up_50 = amdahl(0.5, n_cores)
speed_up_75 = amdahl(0.75, n_cores)
speed_up_90 = amdahl(0.9, n_cores)
speed_up_95 = amdahl(0.95, n_cores)
speed_up_99 = amdahl(0.99, n_cores)

fig, ax = plt.subplots()

ax.plot(n_cores, speed_up_50, label='P = 0.5')
ax.plot(n_cores, speed_up_75, label='P = 0.75')
ax.plot(n_cores, speed_up_90, label='P = 0.9')
ax.plot(n_cores, speed_up_95, label='P = 0.95')
ax.plot(n_cores, speed_up_99, label='P = 0.99')

ax.set_xscale('log')
ax.set_xlabel('Number of cores')
ax.set_ylabel('Speed-up')
ax.set_title('Amdahl\'s Law')
ax.legend()
fig.show()

fig.savefig('amdahl.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

array_size =np.array([int(1e1), int(1e2), int(1e3), int(1e4), int(1e5), int(1e6), int(1e7), int(1e8)])

numpy_time = np.array([1.3e-5, 1.1e-5, 1.1e-5, 1.3e-5, 4.1e-4, 5.0e-4, 2.5e-3, 2.4e-2])
multiprocessing_2_time = np.array([0.16, 0.19, 0.16, 0.17, 0.18, 0.22, 0.41, 5.1])
multiprocessing_4_time = np.array([0.20, 0.21, 0.19, 0.20, 0.35, 0.28, 0.66, 4.2])
multiprocessing_8_time = np.array([0.20, 0.20, 0.21, 0.22, 0.51, 0.56, 0.83, 4.1])

fig, ax = plt.subplots()
ax.plot(array_size, numpy_time, label='numpy', linestyle='solid')
ax.plot(array_size, multiprocessing_2_time, label='multiprocessing 2 cores', linestyle='--')
ax.plot(array_size, multiprocessing_4_time, label='multiprocessing 4 cores', linestyle='-.')
ax.plot(array_size, multiprocessing_8_time, label='multiprocessing 8 cores', linestyle=':')

ax.set_xscale('log')
ax.set_yscale('log')

ax.set_xlabel('Array size')
ax.set_ylabel('Time (s)')
ax.set_title('Performance of numpy and multiprocessing')
ax.legend()

fig.show()
fig.savefig('numpy_comparison.png')
```
