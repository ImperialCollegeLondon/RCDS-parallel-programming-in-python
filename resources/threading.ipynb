```python
import matplotlib.pyplot as plt

n = [1000, 10000, 100000, 1000000, 10000000, 100000000]

time_1 = [0.001, 0.002, 0.02, 0.12, 0.72, 7.2]
time_2 = [0.003, 0.004, 0.02, 0.063, 0.67, 6.6]
time_4 = [0.0028, 0.005, 0.02, 0.061, 0.66, 4.9]
time_8 = [0.005, 0.007, 0.02, 0.058, 0.53, 5.2]

fig, ax = plt.subplots()

ax.plot(n, time_1, label='1 thread', linestyle='--')
ax.plot(n, time_2, label='2 threads', linestyle='-.')
ax.plot(n, time_4, label='4 threads', linestyle = '-')
ax.plot(n, time_8, label='8 threads', linestyle=':')

ax.set(xlabel='n', ylabel='Time (s)')

ax.legend()

ax.set_yscale('log')
ax.set_xscale('log')

fig.savefig("threading_sum.png")
```
